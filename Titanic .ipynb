# Handel Outliers

```python
df
```

```python
Q25 = df['Age'].quantile(0.25)
Q75 = df['Age'].quantile(0.75)
IQR = Q75 - Q25
MIN_Value = Q25 - (1.5*IQR)
MAX_value = Q75 + (1.5*IQR)
```

```python
MAX_value,MIN_Value
```

```python
df.loc[(df['Age']<=MIN_Value) | (df['Age']>=MAX_value)]
```

```python
new_df=df.loc[(df['Age']>=MIN_Value) & (df['Age']<=MAX_value)]
```

```python
new_df.loc[(new_df['Age']<=MIN_Value) | (new_df['Age']>=MAX_value)]
```

```python
new_df=new_df.loc[new_df['Age']<=55]
```

```python
new_df[new_df['Age']>55]
```

```python
new_df
```
